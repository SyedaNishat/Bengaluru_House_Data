```python
import json

meta_path = "AudioMNIST/data/audioMNIST_meta.txt"

with open(meta_path, "r") as f:
    metadata = json.load(f)

print(type(metadata))
```

```python
sample_file = "5_01_12.wav"

parts = sample_file.replace(".wav","").split("_")

digit = parts[0]
speaker = parts[1]

print("Digit:", digit)
print("Speaker:", speaker)
```
