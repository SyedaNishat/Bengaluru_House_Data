```python
import json

meta_path = "AudioMNIST/data/audioMNIST_meta.txt"

with open(meta_path, "r") as f:
    metadata = json.load(f)

print(type(metadata))
```

```python
metadata["01"]
```

```python
sample_file = "5_01_12.wav"

parts = sample_file.replace(".wav","").split("_")

digit = parts[0]
speaker = parts[1]

print("Digit:", digit)
print("Speaker:", speaker)
```

```python
gender = metadata[speaker]["gender"]

print("Gender:", gender)
```

```python
import os
import glob

file_paths = []
digits = []
genders = []

# go through all wav files
audio_files = glob.glob("AudioMNIST/data/*/*.wav")

for file in audio_files:

    filename = os.path.basename(file)

    # split filename
    parts = filename.replace(".wav","").split("_")

    digit = parts[0]
    speaker = parts[1]

    # get gender from metadata
    gender = metadata[speaker]["gender"]

    # store results
    file_paths.append(file)
    digits.append(digit)
    genders.append(gender)
```

```python
print("Total files:", len(file_paths))
print("Example:")
print(file_paths[0], digits[0], genders[0])
```

```python
import pandas as pd

df = pd.DataFrame({
    "file": file_paths,
    "digit": digits,
    "gender": genders
})

df.head()
```
